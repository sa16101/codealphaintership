```python
# Sales and profit by category
category_sales_profit = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()
sns.barplot(data=category_sales_profit, x='Category', y='Profit')
plt.title("Profit by Category")
plt.show()

# Region-wise performance
region_sales_profit = df.groupby('Region')[['Sales', 'Profit']].sum().reset_index()
sns.barplot(data=region_sales_profit, x='Region', y='Profit')
plt.title("Profit by Region")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

category_sales = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()
sns.barplot(data=category_sales, x='Category', y='Sales')
plt.title('Sales by Category')
plt.show()

sns.barplot(data=category_sales, x='Category', y='Profit')
plt.title('Profit by Category')
plt.show()
```

```python
sns.scatterplot(data=df, x='Discount', y='Profit')
plt.title('Discount vs Profit')
plt.show()
```

```python
correlation = df[['Discount', 'Profit']].corr()
print(correlation)
```

```python
state_perf = df.groupby('State')[['Sales', 'Profit']].sum().sort_values(by='Profit')
print(state_perf.head(10))  # Bottom 10 states by profit
```

```python
print(df[df['Profit'] < 0])
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y', errors='coerce')
```

```python
df[df['Order Date'].isna()]
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_sales = df.resample('M', on='Order Date')['Sales'].sum()

monthly_sales.plot(figsize=(12,6), title='Monthly Sales Trend')
plt.ylabel("Sales")
plt.xlabel("Month")
plt.show()
```

```python
# Sales distribution
sns.histplot(df['Sales'], bins=50, kde=True)
plt.title("Sales Distribution")
plt.show()

# Profit distribution
sns.histplot(df['Profit'], bins=50, kde=True, color='green')
plt.title("Profit Distribution")
plt.show()
```

```python
# Sales and profit by category
category_sales_profit = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()
sns.barplot(data=category_sales_profit, x='Category', y='Profit')
plt.title("Profit by Category")
plt.show()

# Region-wise performance
region_sales_profit = df.groupby('Region')[['Sales', 'Profit']].sum().reset_index()
sns.barplot(data=region_sales_profit, x='Region', y='Profit')
plt.title("Profit by Region")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
